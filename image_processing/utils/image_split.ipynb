# Image Split

Run the image split utility from command line
```sh
python image_split.py file_name
```

````python
import cv2
import matplotlib.pyplot as plt

filename = "./images/1_Xg1xaYJ8XVfI28LNBrImyw.webp"
img = cv2.imread(filename)
plt.imshow(img)
````

````python
print(img.shape)
h, w, channels = img.shape
````

````python
half = w // 2

left = img[:,:half]
right = img[:,half:]
````

````python
plt.imshow(left)
````

````python
import os

left_filename = os.path.splitext(filename)[0] + "_left" + os.path.splitext(filename)[1]
cv2.imwrite(left_filename, left)
````
